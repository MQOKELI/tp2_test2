```python
import pandas as pandas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##### Read csv
dataset=pandas.read_csv('C:\\Users\\user\\Downloads\\Informal_Settlement_Programme_-1898186875815614491.csv')

print(dataset.columns)
print(dataset.head())
print(dataset.dtypes)
print(dataset.shape)
```

```python
#First Plot which will be a scatter plot for my graph
plt.scatter(dataset['OBJECTID'].head(10),dataset['WARD_S'].head(10))#OBJECTID,  PROVINCE_C)
plt.title("The RELATIOSHIP BETWEEN OBJECTID and PROVINCE_C")
plt.xlabel("OBJECTID")
plt.ylabel("WARD_S")
plt.show()

#SEcond Plot which will bee a bar graph of my data set but i will extract only two columns and not use the whiole column

plt.bar(dataset['OBJECTID'].head(10),dataset['WATER_SUPP'].head(10), color="Green",alpha=0.7)
plt.xlabel('OBJECTID')
plt.ylabel('WATER_SUPP')
plt.title('BAR GRAPH of OBJECTID vs WATERSUPP')
plt.show()

#Pie chart for a column that i extracted certain values 
SUPPLIMENTS=['SANITATION','WATER_SUPP','REGION ','LATITUDE']
percentage=[30,30,20,20]


plt.figure(figure=(7.7))
plt.pie(percentage,labels=SUPPLIMENTS,autopct='%1.1f%%',startangle=140)
plt.title("SUPPLIMENTS TO THE RESIDENCE OF INFORMAL SETTLEMENTS ")
plt.show()

#ADVANCED PLOTS
sns.kdeplot(dataset['OBJECTID'].head(10),fill=True)#KDE plot using the mentioned  Column which is extracted frm the dataset 
plt.show()

sns.displot(dataset['OBJECTID'])#Distribution plot using same column OBJECTID
plt.show()


#Summary statistics 
print(dataset.describe())
'''mean_values = dataset.mean()
print("The mean of the dataset is "+str(mean_values))

median_values = dataset.median()
print("The meadian of the data set is "+str(median_values))

std_values = dataset.std()
print("The standard deviation of the dataset is "+str(std_values))'''


```
